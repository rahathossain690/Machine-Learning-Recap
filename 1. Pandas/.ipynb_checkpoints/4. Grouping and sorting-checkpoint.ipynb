```python
import pandas as pd
```

```python
df = pd.read_csv('games.csv')
df.head()
```

```python
df.groupby('victory_status')
```

```python
df.groupby('victory_status').mean()
```

```python
df.groupby('victory_status').apply(lambda df2: df2.turns)
```

```python
# groupby basically is making a group based on the attribute keeping all the data
```

```python
df.groupby(['victory_status', 'winner']).apply(lambda x: x.white_rating.max())
```

```python
df.groupby(['victory_status']).turns.agg([min, max])
# multiple-group-by-dependency makes the resulting DataFrame multi-indexed. 
```

```python
df.sort_values(by=['turns', 'winner', 'victory_status'])
```
