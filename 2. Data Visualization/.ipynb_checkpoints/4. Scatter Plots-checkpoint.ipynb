```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
df = pd.read_csv('spotify.csv', index_col='Date')
df.tail()
```

```python
sns.scatterplot(x=df['Shape of You'], y=df['Despacito'], hue=df['HUMBLE.'])
```

```python
sns.lmplot(x="Shape of You", y="Despacito", hue="HUMBLE.", data=df)
```

```python
sns.lmplot(x="Shape of You", y="Despacito", data=df)
# tries to make a regression line
```

```python
sns.swarmplot(x=df['Shape of You'],
              y=df['Despacito'])
```
