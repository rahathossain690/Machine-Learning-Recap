```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
df = pd.read_csv('fifa.csv', index_col="Date")
df.head()
```

```python
sns.distplot(a=df['ARG'], kde=False)
```

```python
sns.kdeplot(data=df['ARG'], shade=True)
```

```python
sns.jointplot(x=df['ARG'], y=df['GER'], kind="kde")
```

```python
sns.distplot(a=df['ARG'], kde=False)
sns.distplot(a=df['BRA'], kde=False)

plt.title("Comparison between Bra and Arg")
plt.legend(['Brazil', 'Argentina'])
```

```python
sns.kdeplot(data=df['ARG'], shade=True)
sns.kdeplot(data=df['BRA'], shade=True)
sns.kdeplot(data=df['GER'], shade=True)
```
