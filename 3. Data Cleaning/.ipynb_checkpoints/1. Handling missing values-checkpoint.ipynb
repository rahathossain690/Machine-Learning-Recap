```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('spotify.csv')
df.head()
```

```python
# counting missing data
df.isnull().sum()
```

```python
total_cell = np.product(df.shape)
total_miss = df.isnull().sum().sum() # first sum returns for all colms, second one merges all
total_miss
```

```python
df.dropna() # drops all the missing valued row
```

```python
df.dropna(axis=1) # drop columns with missing values
```

```python
df.loc[0:5, 'Shape of You': 'HUMBLE.'].fillna(0) # filling missing data with something
```

```python
# replace all NA's the value that comes directly after it in the same column, 
# then replace all the remaining na's with 0
df.fillna(method='bfill', axis=0).fillna(0)
```
